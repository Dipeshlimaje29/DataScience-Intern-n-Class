# Scrapping The flight ticket price In interval of 1day, 15 day , 30 day

```python
# ASSUMPTIONS
# 1] I have assumed some location for departure and arrival
# 2] I have kept all classes mixed (Economy, business, general)
# 3] I have scrapped the data by keeping 3 Jan 2023 as the reference date and assumed intervals of 1 day, 15 days, and 30 days from 3jan2023
# 4] I have scrapped the data from one site Yatra.com
# 5] I have also kept the price range unlimited
# 6] I have kept stops from non-stop to a maximum 3 stop 
# 7] I have kept all 24 hrs flights not particular timing 
```

## Delhi-Mumbai 1 day interval

```python
# lets import nec library
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
import re
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
#lets first connect webdriver
driver=webdriver.Chrome('chromedriver')

#maximize the automated window
driver.maximize_window()
```

```python
#open the link 
url='https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=04%2F01%2F2023&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=71911160047'
driver.get(url)
```

```python
#5000 time we scroll down for getting atleast 500 comments

for _ in range(50000):

    driver.execute_script("window.scrollBy(0,500000)")
```

```python
#filght_name
filght_name=[]
names = driver.find_elements(By.XPATH,'//span[@class="i-b text ellipsis"]')

for name in names[0:500]:

    filght_name.append(name.text)
```

```python
#dep_time
dep_time=[]
names = driver.find_elements(By.XPATH,'//div[@class="i-b pr"]')

for name in names[0:500]:

    dep_time.append(name.text)
```

```python
#arr_time
arr_time=[]
names = driver.find_elements(By.XPATH,'//div[@class="i-b pdd-0 text-left atime col-5"]/p[1]')

for name in names[0:500]:

    arr_time.append(name.text.split('\n')[0])
```

```python
#tot_time
tot_time=[]
names = driver.find_elements(By.XPATH,'//div[@class="stop-cont pl-13"]/p')

for name in names[0:500]:

    tot_time.append(name.text)
```

```python
#tot_stop
tot_stop=[]
names = driver.find_elements(By.XPATH,'//div[@class="stop-cont pl-13"]/div/span')

for name in names[0:500]:

    tot_stop.append(name.text)
```

```python
#price
price=[]
names = driver.find_elements(By.XPATH,'//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')

for name in names[0:500]:

    price.append(name.text)
```

```python
#creating dataframe 

DF1=pd.DataFrame({'Flight Name':filght_name,'Depart Location':dep_name,'Arrival Location':arr_name, 'Depart Time':dep_time ,'Arrival Time':arr_time , 'Total Stop':tot_stop ,'Total Time':tot_time,'Price':price,'Date of Journey':'4/1/2023'})
DF1
```

```python
#open the link 
url='https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=04%2F01%2F2023&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=71911160047'
driver.get(url)
```

```python
# entering Carwow
search_job=driver.find_element(By.XPATH,'/html/body/section[1]/section/div/div/form/ul[2]/li[2]/div/input')
search_job.send_keys('nagpur')
```

```python
# click search button
search_btn=driver.find_element(By.XPATH,'/html/body/section[1]/section/div/div/form/div[1]/div/button/span')
search_btn.click()
```

```python
#5000 time we scroll down for getting atleast 500 comments

for _ in range(50000):

    driver.execute_script("window.scrollBy(0,500000)")
```

```python
#filght_name
filght_name=[]
names = driver.find_elements(By.XPATH,'//span[@class="i-b text ellipsis"]')

for name in names[0:500]:

    filght_name.append(name.text)
    
    
#dep_name
dep_name=[]
names = driver.find_elements(By.XPATH,'//div[@class="i-b col-4 no-wrap text-right dtime col-3"]/p')

for name in names[0:500]:

    dep_name.append(name.text)
    
    
#arr_name
arr_name=[]
names = driver.find_elements(By.XPATH,'//div[@class="i-b pdd-0 text-left atime col-5"]/p[2]')

for name in names[0:500]:

    arr_name.append(name.text)
    

#dep_time
dep_time=[]
names = driver.find_elements(By.XPATH,'//div[@class="i-b pr"]')

for name in names[0:500]:

    dep_time.append(name.text)
    
    
    
#arr_time
arr_time=[]
names = driver.find_elements(By.XPATH,'//div[@class="i-b pdd-0 text-left atime col-5"]/p[1]')

for name in names[0:500]:

    arr_time.append(name.text.split('\n')[0])
    
    
    
#tot_time
tot_time=[]
names = driver.find_elements(By.XPATH,'//div[@class="stop-cont pl-13"]/p')

for name in names[0:500]:

    tot_time.append(name.text)
    
    
    
#tot_stop
tot_stop=[]
names = driver.find_elements(By.XPATH,'//div[@class="stop-cont pl-13"]/div/span')

for name in names[0:500]:

    tot_stop.append(name.text)
    
    
    
#price
price=[]
names = driver.find_elements(By.XPATH,'//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')

for name in names[0:500]:

    price.append(name.text)
```

```python
#creating dataframe 

DF21=pd.DataFrame({'Flight Name':filght_name,'Depart Location':dep_name,'Arrival Location':arr_name, 'Depart Time':dep_time ,'Arrival Time':arr_time , 'Total Stop':tot_stop ,'Total Time':tot_time,'Price':price,'Date of Journey':'19/1/2023'})
DF21
```

```python
DF21.to_csv('flight_data_21.csv')
```
