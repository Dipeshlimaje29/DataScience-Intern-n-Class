## 7. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
# lets import nec library
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
import re
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
#lets first connect webdriver
driver=webdriver.Chrome('chromedriver')

#maximize the automated window
driver.maximize_window()
```

```python
#open the link 
url='https://www.digit.in/'
driver.get(url)
```

```python
# click search button
search_btn=driver.find_element(By.XPATH,'/html/body/div[3]/div/div[2]/div[2]/div[4]/ul/li[9]/a')
search_btn.click()
```

```python
# URLs for Gaming Laptop
product_URL = []
start=0
end=1
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//div[@class="TopNumbeHeading"]/div/a')
    for i in url:
        product_URL.append(i.get_attribute('href'))
```

```python
product_URL
```

```python
# Windows of gaming laptop
win=[]
for i in product_URL:
    driver.get(i)
    time.sleep(4)
    try:
        price=driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[3]/div[2]/div[3]/div/ul/li[1]/div/p[2]/strong')
        win.append(price.text)
    except NoSuchElementException:
        win.append('-')
```

```python
win
```

```python
# Display of laptop
display=[]
for i in product_URL:
    driver.get(i)
    time.sleep(4)
    try:
        price=driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[3]/div[2]/div[3]/div/ul/li[2]/div/p[2]/strong')
        display.append(price.text)
    except NoSuchElementException:
        display.append('-')
```

```python
display
```

```python
# Processor of laptop
pro=[]
for i in product_URL:
    driver.get(i)
    time.sleep(4)
    try:
        price=driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[3]/div[2]/div[3]/div/ul/li[3]/div/p[2]/strong')
        pro.append(price.text)
    except NoSuchElementException:
        pro.append('-')
```

```python
pro
```

```python
# Ram and Storage Avialble for Laptop
memory=[]
for i in product_URL:
    driver.get(i)
    time.sleep(4)
    try:
        price=driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[3]/div[2]/div[3]/div/ul/li[4]/div/p[2]/strong')
        memory.append(price.text)
    except NoSuchElementException:
        memory.append('-')
```

```python
memory
```

```python
# Name of the Laptop
name=[]
for i in product_URL:
    driver.get(i)
    time.sleep(4)
    try:
        price=driver.find_element(By.XPATH,'/html/body/div[3]/div[2]/div[2]/h1')
        name.append(price.text)
    except NoSuchElementException:
        name.append('-')
```
